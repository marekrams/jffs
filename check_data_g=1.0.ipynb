```python
import glob
import os
from operator import itemgetter
from itertools import groupby
```

```python
g = 1
v, Q = 1, 1
#
glob_path = os.path.join(os.path.abspath(""), "./results_fermions/g=1.0000/**/results.npy")
fnames = glob.glob(glob_path, recursive=True)
fnames = [x.split("/g=1.0000/")[1] for x in fnames]
```

```python
def parse_string(txt):
    res = {}
    for x in txt.split("/"):
        if "=" in x:
            y = x.split("=")
            res[y[0]] = float(y[1]) if "." in y[1] or 'e' in y[1] else int(y[1])
        elif 'site' in x:
            res['method'] = x
    return res
```

```python
data = [parse_string(txt)  for txt in fnames]
data = sorted(data, key=itemgetter('m'))
datam = {k: list(v) for k, v in groupby(data, key=itemgetter('m'))}
```

```python
for k, v in datam.items():
    for w in v:
        if w['a'] <= 0.25:
            print(w)
    print("--------------")
```
